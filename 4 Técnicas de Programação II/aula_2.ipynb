## Pandas Plot

```python
import pandas as pd
```

```python
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"

df = pd.read_csv(url)
```

```python
df.shape
```

```python
df.columns
```

```python
df.head()
```

## Gráfico de Linhas

```python
df.location.unique()
```

```python
df1 = df[df['location']=='Brazil']
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.plot(
    kind='line',
    x='date',
    y='total_cases',
    rot=90,
    title="Agregado de Casos da Covid-19 no Brasil",
    xlabel="Datas",
    ylabel="Qtd de casos",
    style='--r',
    figsize=(10,5),
    grid=True
).get_figure().savefig('grafico_bacana.png')
```

```python
df1.loc[:,['date','total_cases','total_deaths']].plot(
    kind='line',
    logy=True,
    grid=True,
    style={
        "total_cases":"--r",
        "total_deaths":",k"
        }
);
```

## Gráficos de Barra

```python
df2 = df[df['continent']=='South America'].groupby('location')['total_cases'].max()
```

```python
df2.plot(
    kind='bar',
    figsize=(11,5)
)
```

```python
df3 = df[df['continent']=='South America'].groupby('location')['total_cases','total_deaths'].max()
```

```python
df3.plot(
    kind='bar',
    figsize=(11,5),
    logy=True
)
```

```python
df3.plot(
    kind='barh',
    figsize=(11,5),
    logx=True
)
```

```python
df3.plot(
    kind='barh',
    figsize=(11,5),
    # logx=True,
    stacked=True
)
```

# Gráfico de Dispersão

```python
df3
```

```python
df3 = df3.reset_index(level=0)
```

```python
df3.plot(
    kind='scatter',
    x='total_cases',
    y='total_deaths',
    cmap="plasma",
    c=df3.index
)
```
