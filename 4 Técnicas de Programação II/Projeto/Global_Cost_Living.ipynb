## **Descrição do dataset:**

Este conjunto de dados, contém informações sobre o custo de vida em quase 5.000 cidades em todo o mundo. 
**Custo de vida (ou ICV)** é um indicador calculado sobre a soma dos preços médios pagos por bens e serviços em um determinado lugar.
o custo de vida é um indicador de quão caro é morar em um determinado lugar, mantendo um determinado padrão de vida.
Esse índice deve ser examinado levando sempre em consideração a média salarial do mesmo lugar.

Link do dataset: [Global living cost](https://www.kaggle.com/datasets/mvieira101/global-cost-of-living)

## **Caso Beltrano**



**Beltrano** foi *trocado* por sua esposa (🤘) e quer um país que tenha a ***bebida mais forte*** para que ele possa esquecer sua ex-mulher. Ele começou pesquisando os maiores preços de *Vinho, Cervejas importadas e Cervejas locais*. Pois segundo ele, quanto **maior** o preço que ele pagasse, mais **forte** a bebida seria. Para isso, ele escolheu comprar um combo contendo cada uma dessas bebidas.

```python
# Descobrindo o valor da Vinho nos países
df_por_pais = df
df_por_pais[['country','x24']].sort_values(by='x24', ascending=False).head(10)
```

```python
# Descobrindo o valor da Cerveja local nos países
df_por_pais[['country','x25']].sort_values(by='x25', ascending=False).head(10)
```

```python
#Descobrindo o valor da Cerveja importada nos países
df_por_pais[['country','x26']].sort_values(by='x26', ascending=True).head(10)
```

```python
# Calculando o preço para Esquecer a Ex Mulher.
df_bebidas_por_pais = df_por_pais[['country', 'x24', 'x25', 'x26']].copy()
df_bebidas_por_pais.rename(columns={'country': 'País', 
                                    'x24': 'Garrafa de Vinho',
                                    'x25': 'Cerveja Local',
                                    'x26': 'Cerveja Importada'}, inplace=True)

df_bebidas_por_pais['Preço p/ Esquecer'] = df_bebidas_por_pais.iloc[:, 1:].sum(axis=1)
display(df_bebidas_por_pais)
```

```python
# Matplotlib
plt.figure(figsize=(20, 10))
plt.scatter(df_bebidas_por_pais['País'], df_bebidas_por_pais['Preço p/ Esquecer'])
plt.xlabel("País")
plt.ylabel("Preço em Reais")
plt.title("Preço para esquecer a ex mulher")
plt.show()
```

**Sicrano** está estudando na Ada e gostaria de ter noção do custo de vida de alguns países onde observou que há grande oferta para devs. Para isso, com base nos dados de custo de vida de um dataset, calculou o valor médio dos custos em cada país e, em seguida, elencou alguns itens que considera essenciais à realização da sua análise.

## Plotly - Mapa Coroplético

```python
import plotly.graph_objects as go
import pandas as pd

df_localizacao_paises = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
df_localizacao_paises
```

```python
# df_localizacao_paises = df_localizacao_paises.drop('GDP (BILLIONS)', axis=1)
df_localizacao_paises = df_localizacao_paises.rename(columns = {'COUNTRY': 'country'})
df_localizacao_paises
```

```python
df[['country', 'x1']]
```

```python
df_mapa = pd.merge(df_localizacao_paises[['country', 'CODE']], df, on = 'country')
df_mapa
```

```python
def mostrar_mapa_caracteristica(caracteristica: str):
  fig = go.Figure(data=go.Choropleth(
                    locations = df_mapa['CODE'],
                    z = df_mapa[caracteristica],
                    text = df_mapa['country'],
                    colorscale = 'Blues',
                    autocolorscale=False,
                    reversescale=True,
                    marker_line_color='darkgray',
                    marker_line_width=0.5,
                    colorbar_tickprefix = '$',
                    colorbar_title = 'US$'
  ))

  fig.update_layout(
      geo=dict(
          showframe=False,
          showcoastlines=False,
          projection_type='equirectangular'
      )
  )

  fig.show()
```
