```python
import pandas as pd
import matplotlib.pyplot as plt

url_path = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'

df = pd.read_csv(url_path)

mask = df['location'] == 'Brazil'
dados = df[mask][['date', 'new_deaths']]
dados = dados.dropna()

dados2 = dados.copy()
dados2['date'] = pd.to_datetime(dados2['date'])

# Ponto máximo
ponto_maximo = dados2[dados2['new_deaths'] == dados2['new_deaths'].max()]

plt.figure(figsize = (20, 6))
plt.plot(dados2.set_index('date')['new_deaths'])
plt.title('Evolução de óbitos confirmados no Brasil')
plt.ylabel('Número de óbitos')
plt.xlabel('Data')
plt.show()
```

```python
plt.plot(df_br['new_deaths'])
plt.plot(df_br['new_deaths'].max())
```

```python
df1 = df[df['location']=='Brazil']
df1 = df1.reset_index(drop=True)
df1.head()
```

```python
x = df1['new_deaths']

plt.figure(figsize=(19,7))
plt.plot(x, label='Óbitos diários')
plt.vlines(
    x=x.idxmax(),
    ymin=x.min(),
    ymax=x.max(),
    linestyles='--',
    linewidth=3,
    label='Limiar do Pico de óbitos'
)
plt.plot(x.idxmax(), x.max(), 'x', color='r',  label='Pico de óbitos')
plt.ylabel('Número de óbitos')
plt.xlabel('Data')
plt.legend()
```
