### Escrevendo dados de um arquivo

```python
df.to_excel('arq4.xlsx', index=False)
```

Exercício 

Leia o arquivo `aluno3.csv`, apenas as colunas RA, Prova_1, Prova_2, Prova_3, Prova_4 realise a médias das notas da prova e salve em um arquivo excel (xlsx) as colunas RA e Média.


```python
import pandas as pd

df = pd.read_csv("dados/alunos3.csv", sep = ";", decimal = ',', usecols=['RA', 'Prova_1', 'Prova_2', 'Prova_3', 'Prova_4'])
display(df)

df['media_final'] = df[[f'Prova_{i + 1}' for i in range(4)]].mean(axis = 1).round(2)
print("\nCriação da coluna média: ")
display(df)

print("\nExport final: ")
display(df[['RA', 'media_final']])
df[['RA', 'media_final']].to_excel('notas.xlsx', index = False)
```
