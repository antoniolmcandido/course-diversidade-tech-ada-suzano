```python
# Bibliotecas
import os
import random
from google.colab import output
```

```python
def escolher_palavra():  
  with open('frutas.txt', 'r') as arquivo:
    frutas = [linha.replace('\n', '') for linha in arquivo]

  palavra = random.choice(frutas)
  return palavra
```

```python
def representar(palavra):
  representacao = list(len(palavra) * '-')
  return representacao
```

```python
def substituir(palavra, representacao, letra):  
  acerto = False
  for idx, char in enumerate(palavra):
    if char == letra:
      representacao[idx] = letra
      acerto = True

  return acerto
```

```python
def main():  
  erro = 0    
  letras_usadas = []  
  palavra = escolher_palavra()
  representacao = representar(palavra)

  while True:
    output.clear()

    print(erros.get(erro))   
    if erro == 6 : 
      print('Que pena, você perdeu :(')
      print('A palavra era:', palavra)          
      print('Letras usadas: ', letras_usadas)
      break

    if '-' not in representacao:
      print('PARABÉNS, você ganhou :)')
      print('A palavra era:', palavra)
      print('Letras usadas: ', letras_usadas)
      break

    print('Palavra secreta: ', end='')
    print(''.join(representacao))
    print('Letras usadas: ', letras_usadas)
    
    letra = input('Digite uma letra que não foi utilizada: ')
    while letra in letras_usadas:
      letra = input('Digite uma letra que não foi utilizada: ')
    
    letras_usadas.append(letra)
    acerto = substituir(palavra, representacao, letra)

    if acerto: pass
    else: erro += 1
```
